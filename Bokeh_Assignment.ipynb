Q.No-02    What are glyphs in Bokeh, and how can you add them to a Bokeh plot? Explain with an example.

Ans :-

In Bokeh, glyphs are visual markers or symbols used to represent data points in a plot. They are fundamental building blocks for creating various types of visualizations, such as scatter plots, line plots, bar charts, and more. Bokeh provides a wide range of glyphs, including circles, squares, triangles, lines, bars, and more, allowing you to customize and present your data in different ways.

To add glyphs to a Bokeh plot, you need to use the `figure` function from the `bokeh.plotting` module and then call the glyph methods available for the plot object. Each glyph method corresponds to a specific type of glyph, and you can use them to specify the data and visual properties of the glyphs.

    Here's an example of how to add glyphs to a Bokeh plot:

````python
from bokeh.plotting import figure, show, output_notebook, output_file
from bokeh.io.export import export_png

# Sample data 
x = [1, 2, 3, 4, 5]
y = [6, 7, 2, 4, 5]
````

**`Circle`**

````python
# Create a new plot with specified title and axis labels
Circle = figure(title="Glyphs Example", x_axis_label="X-axis", y_axis_label="Y-axis")

# Add circle glyphs to the plot
Circle.circle(x, y, size=10, color="navy", alpha=0.8, legend_label="Circles")

output_notebook()
show(Circle)

export_png(Circle, filename="Circle.png")

from IPython.display import Image
Image(filename='Circle.png')
````

**`Triangle`**

````python
Triangle = figure(title="Glyphs Example", x_axis_label="X-axis", y_axis_label="Y-axis")

# Add triangle glyphs to the plot
Triangle.triangle(x, y, size=12, color="firebrick", alpha=0.6, legend_label="Triangles")

output_notebook()
show(Triangle)

export_png(Triangle, filename="Triangle.png")

from IPython.display import Image
Image(filename='Triangle.png')
````

**`Line`**

````python
Line = figure(title="Glyphs Example", x_axis_label="X-axis", y_axis_label="Y-axis")

# Add line glyphs to the plot
Line.line(x, y, line_width=2, color="green", alpha=0.5, legend_label="Line")

output_notebook()
show(Line)

export_png(Line, filename="Line.png")

from IPython.display import Image
Image(filename='Line.png')
````

**`Square`**

````python
Square = figure(title="Glyphs Example", x_axis_label="X-axis", y_axis_label="Y-axis")

# Add square glyphs to the plot
Square.square(x, y, size=8, color="orange", alpha=0.7, legend_label="Squares")

output_notebook()
show(Square)

export_png(Square, filename="Square.png")

from IPython.display import Image
Image(filename='Square.png')
````

Q.No-03    How can you customize the appearance of a Bokeh plot, including the axes, title, and legend?

Ans :-

````python
from bokeh.plotting import figure, show
from bokeh.models import ColumnDataSource
from bokeh.io import output_notebook

# Example data
x = [1, 2, 3, 4, 5]
y1 = [10, 20, 15, 30, 25]
y2 = [5, 15, 10, 25, 20]

# Create a ColumnDataSource to hold the data
source = ColumnDataSource(data={'x': x, 'y1': y1, 'y2': y2})

# Create a figure object
Q_No_03 = figure(title="Customized Bokeh Plot", x_axis_label="X-axis", y_axis_label="Y-axis")

# Customize the appearance of the axes
Q_No_03.xaxis.axis_label_text_color = "red"
Q_No_03.xaxis.axis_label_standoff = 15
Q_No_03.xaxis.axis_line_color = "blue"
Q_No_03.xaxis.major_label_text_color = "green"

Q_No_03.yaxis.axis_label_text_font_style = "italic"
Q_No_03.yaxis.axis_line_color = "orange"
Q_No_03.yaxis.major_label_text_color = "purple"

# Customize the appearance of the plot title
Q_No_03.title.text_color = "black"
Q_No_03.title.text_font_size = "16pt"
Q_No_03.title.align = "center"

# Create glyphs and legends for each series
Q_No_03.line(x='x', y='y1', source=source, legend_label="Series 1", line_color="red", line_width=2)
Q_No_03.circle(x='x', y='y1', source=source, legend_label="Series 1", size=8, fill_color="red")

Q_No_03.line(x='x', y='y2', source=source, legend_label="Series 2", line_color="blue", line_width=2)
Q_No_03.circle(x='x', y='y2', source=source, legend_label="Series 2", size=8, fill_color="blue")

# Show the plot with the legend
show(Q_No_03)

export_png(Q_No_03, filename="Q_No_03.png")

from IPython.display import Image
Image(filename='Q_No_03.png')
````

Q.No-04    What is a Bokeh server, and how can you use it to create interactive plots that can be updated in real time?

Ans :-

Bokeh is a Python library for creating interactive data visualizations. Bokeh Server is a component of the Bokeh library that allows you to build and deploy interactive web applications with real-time updating capabilities. It enables you to create interactive plots, dashboards, and applications that can be accessed through a web browser.

Bokeh Server works by establishing a two-way communication between the Python code running on the server and the web browser. This allows you to create plots and visualizations in Python, and as the data changes on the server-side, the changes are automatically pushed to the client-side (browser), updating the visualization in real time. This is particularly useful when you want to provide users with a dynamic and interactive experience while working with data.

Here's a general outline of how you can use Bokeh Server to create interactive plots that update in real-time -

1. **Install Bokeh**: If you haven't already, install Bokeh using the following command in a Jupyter Notebook cell:

2. **Import Required Libraries**:

   In a Jupyter Notebook cell, import the necessary libraries:

````python
from bokeh.plotting import figure, show, output_notebook
from bokeh.models import ColumnDataSource
from bokeh.io import push_notebook
import numpy as np
from IPython.display import display
import time
````

3. **Initialize Notebook Output** -

   To display Bokeh plots directly in the notebook, use the following command:

````python
output_notebook()
````

4. **Create a Bokeh Figure and Data Source** -

   In a notebook cell, create the figure and data source as before:

````python
plot = figure(width=400, height=400)
source = ColumnDataSource(data={'x': np.random.random(10), 'y': np.random.random(10)})
````

5. **Define Callback Function** -

   Define the callback function to update the plot data source:

````python
def update_data():
       new_data = {'x': np.random.random(10), 'y': np.random.random(10)}
       source.data = new_data
       push_notebook()
````

6. **Add Plot Elements** -

   Add plot elements to the figure:

````python
plot.circle('x', 'y', source=source, size=10)
````

7. **Display the Initial Plot** -

   Display the initial plot using the `show()` function and the `push_notebook()` function to enable updates:

````python
plot_handle = show(plot, notebook_handle=True)

export_png(plot, filename="plot.png")

from IPython.display import Image
Image(filename='plot.png')
````

8. **Run the Callback** -

   Run the callback in a loop to update the plot data source at regular intervals. Use the `display()` function to update the notebook cell's output:

````python
from IPython.display import display
import time

for _ in range(10):  # Update the plot 10 times
   update_data()
   display(plot)
   time.sleep(1)  # Wait for 1 second
````

This code will update the plot's data source and display the updated plot in the notebook cell every 1 second for a total of 10 updates.

Remember that this is just a simplified example for demonstration. You can customize the behavior and appearance of the plot, integrate user interactions, or even use widgets to control the updates and interactivity.

Q.No-05    How can you embed a Bokeh plot into a web page or dashboard using Flask or Django?

Ans :-

### **Using Flask :**

To embed a Bokeh plot into a web page or dashboard using Flask or Django, you can use the `bokeh.embed.components()` function. This function returns a script and a div tag that can be used to embed the plot in an HTML page.

The following code shows how to embed a Bokeh plot in a Flask app:

1. Install the required packages using pip:
   
   ```
   pip install Flask bokeh
   ```

2. Create a file named `app.py` with the following code:

````python
# this in app.py
from flask import Flask, render_template
from bokeh.plotting import figure, show
from bokeh.embed import components

app = Flask(__name__)

@app.route('/')
def index():
    # Create a Bokeh plot
    plot = figure()
    plot.circle([1, 2, 3, 4, 5], [6, 7, 2, 4, 5])

    # Generate the JavaScript and HTML components for embedding
    script, div = components(plot)

    return render_template('index.html', script=script, div=div)

if __name__ == '__main__':
    app.run()
````

3. Create a folder named `templates` in the same directory as your `app.py` file, and inside that folder, create a file named `index.html` with the following content:

```html
<!DOCTYPE html>
<html>
<head>
    <title>Bokeh Plot</title>
    <!-- Include Bokeh JS library -->
    <script src="https://cdn.bokeh.org/bokeh/release/bokeh.min.js"></script>
</head>
<body>
    <h1>Bokeh Plot</h1>
    <!-- Insert Bokeh plot components -->
    {{ script|safe }}
    {{ div|safe }}
</body>
</html>
```
